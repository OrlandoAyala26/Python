# Problema 1



```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el centro y radio de la bola abierta
centro = [0,0]
radio = 1

# Generamos una malla de puntos en el cuadrado [-2, 2] x [-2, 2]
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

# Uso de meshgrid
# https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html#numpy.meshgrid
xx, yy = np.meshgrid(x, y)

# Calculamos la norma L1 de cada punto de la malla e imponemos la condición de centrado
norma1 = np.abs(xx-centro[0]) + np.abs(yy-centro[1])

# Graficamos los puntos dentro de la bola abierta
# Uso de contour
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html
plt.contour(xx, yy, (norma1 < radio))

# Establecemos los límites de los ejes y mostrar el gráfico
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title("Bola abierta B(0,1) con norma L 1")
plt.xlabel("Irving Orlando Ayala Iturbe")
plt.show()
```

# Prueba en 3D

```python
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D

# Define el centro de la bola y el radio
centro = [0, 0, 0]
radio = 1

# Crea una malla de puntos en la bola
theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, np.pi, 100)
x = radio * np.outer(np.cos(theta), np.sin(phi))
y = radio * np.outer(np.sin(theta), np.sin(phi))
z = radio * np.outer(np.ones(100), np.cos(phi))

# Calcula la norma euclidiana de cada punto
normas = np.sqrt(x**2 + y**2 + z**2)

# Crea la figura y el objeto de ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grafica los puntos dentro de la bola abierta
ax.scatter(x[normas < radio], y[normas < radio], z[normas < radio], c='b')

# Establece los límites de los ejes
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)

# Muestra el gráfico
plt.show()
```
